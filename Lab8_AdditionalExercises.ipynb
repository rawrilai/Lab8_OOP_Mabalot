```python
class Book:
    def __init__(self, title, author, available=True):
        self.title = title
        self.author = author
        self.available = available

class Library:
    def __init__(self):
        self.books = []

    def add_book(self, book):
        self.books.append(book)

    def display_books(self):
        print("\nAvailable Books:")
        for book in self.books:
            status = "Available" if book.available else "Not Available"
            print(f"{book.title} by {book.author} - {status}")

    def borrow_book(self, title):
        for book in self.books:
            if book.title == title and book.available:
                book.available = False
                return f"You have borrowed '{book.title}' by {book.author}."
        return "Sorry, that book is not available."

    def return_book(self, title):
        for book in self.books:
            if book.title == title and not book.available:
                book.available = True
                return f"You have returned '{book.title}' by {book.author}."
        return "Sorry, that book was not borrowed."
    
lib = Library()
lib.add_book(Book("A Room with a View", "E.M. Forster", True))
lib.add_book(Book("White Nights", "Fyodor Dostoevsky", False))
lib.add_book(Book("The Great Gatsby", "F. Scott Fitzgerald", True))
lib.add_book(Book("Pride and Prejudice", "Jane Austen", False))
lib.add_book(Book("The Count of Monte Cristo", "Alexandre Dumas", False))

while True:
    print("\n1. Display Books\n2. Borrow Book\n3. Return Book\n4. Exit")
    choice = input("Choice: ")

    if choice == "1":
        lib.display_books()
    elif choice == "2":
        title = input("Enter title to borrow: ")
        print(lib.borrow_book(title))
    elif choice == "3":
        title = input("Enter title to return: ")
        print(lib.return_book(title))
    elif choice == "4":
        print("Exiting program.")
        break
    else:
        print("Invalid choice.")
```

```python
class Employee:
    def __init__(self, name, hourly_rate, hours_worked):
        self.name = name
        self.hourly_rate = hourly_rate
        self.hours_worked = hours_worked

    def compute_salary(self):
        if self.hours_worked > 160:
            overtime_hours = self.hours_worked - 160
            overtime_pay = overtime_hours * self.hourly_rate * 1.5
            base_salary = 160 * self.hourly_rate
        else:
            base_salary = self.hours_worked * self.hourly_rate
            overtime_pay = 0

        return base_salary, overtime_pay


class Payroll:
    def __init__(self):
        self.employees = []

    def add_employee(self, emp):
        self.employees.append(emp)

    def display_payroll(self):
        print("\nPayroll Summary:")
        for emp in self.employees:
            base, overtime = emp.compute_salary()  # FIXED HERE
            print(f"\nName: {emp.name}")
            print(f"Hours Worked: {emp.hours_worked}")
            print(f"Base Salary: {base:.2f}")
            print(f"Overtime Pay: {overtime:.2f}")


payroll = Payroll()

payroll.add_employee(Employee("Cloud", 1500, 170))
payroll.add_employee(Employee("Yana", 1200, 150))
payroll.add_employee(Employee("Waffle", 2000, 180))
payroll.add_employee(Employee("Imu", 1800, 160))

n = int(input("Enter number of employees to add: "))

for i in range(n):
    print(f"\nEmployee {i+1}:")
    name = input("Name: ")
    rate = float(input("Hourly Rate: "))
    hours = float(input("Hours Worked: "))

    payroll.add_employee(Employee(name, rate, hours))

payroll.display_payroll()
```

```python
class Product:
    def __init__(self, name, price, stock):
        self.name, self.price, self.stock = name, price, stock


class Cart:
    def __init__(self, products):
        self.products = products
        self.items = {}

    def add(self, name, qty):
        p = next((x for x in self.products if x.name.lower() == name.lower()), None)

        if not p:
            print("Invalid product.")
            return
        if qty > p.stock:
            print("Not enough stock.")
            return

        p.stock -= qty
        self.items[name] = self.items.get(name, 0) + qty
        print("Added successfully.")

    def remove(self, name, qty):
        if name not in self.items:
            print("Not in cart.")
            return
        if qty > self.items[name]:
            print("Too many to remove.")
            return

        self.items[name] -= qty
        if self.items[name] == 0:
            del self.items[name]

        print("Removed successfully.")

    def total(self):
        return sum(
            qty * p.price
            for name, qty in self.items.items()
            for p in self.products
            if p.name == name
        )

    def show(self):
        print("\nCart:")
        for name, qty in self.items.items():
            price = next(p.price for p in self.products if p.name == name)
            print(f"{name} x{qty} = {qty * price}")
        print("Total:", self.total())

products = [
    Product("Apple", 100, 50),
    Product("Milk", 120, 30),
    Product("Bread", 90, 25)
]

cart = Cart(products)

while True:
    print("\n1-See Available Products 2-Add Product 3-Remove Product 4-View Cart 5-Checkout")
    choice = input("Choice: ")

    if choice == "1":
        print("\nAVAILABLE PRODUCTS:")
        for p in products:
            print(f"{p.name} - ₱{p.price} - Stock: {p.stock}")

    elif choice == "2":
        name = input("Product name: ")
        qty = int(input("Quantity: "))
        cart.add(name, qty)
    elif choice == "3":
        name = input("Product name: ")
        qty = int(input("Quantity to remove: "))
        cart.remove(name, qty)

    elif choice == "4":
        cart.show()

    elif choice == "5":
        print("Checkout complete.")
        break
    else:
        print("Invalid choice.")
```
